```python
# Define a function to determine the 'shift' value based on 'hour'
def determine_shift(hour):
    if hour >= 4 and hour <= 9:
        return 'Shift 3'
    elif hour >= 13 and hour <= 17:
        return 'Shift 1'
    elif hour >= 21 and hour <= 23:
        return 'Shift 2'

# Apply the function to create the 'shift' column based on 'hour'
df_2023['shift'] = df_2023['hour'].apply(determine_shift)
```

```python
plt.figure(figsize=(6,6))
plt.title('Boxplot to detect outliers for humidity (%)', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
sns.boxplot(x=df_2023['humidity (%)'])
plt.show()
```

```python
# Determine the number of rows containing outliers 
### YOUR CODE HERE ###

# Compute the 25th percentile value in `humidity`
percentile25 = df_2023['humidity (%)'].quantile(0.25)

# Compute the 75th percentile value in `tenure`
percentile75 = df_2023['humidity (%)'].quantile(0.75)

# Compute the interquartile range in `humidity`
iqr = percentile75 - percentile25

# Define the upper limit and lower limit for non-outlier values in `humidity`
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)

# Identify subset of data containing outliers in `humidity`
outliers = df_2023[(df_2023['humidity (%)'] > upper_limit) | (df_2023['humidity (%)'] < lower_limit)]

# Count how many rows in the data contain outliers in `humidity`
print("Number of rows in the data containing outliers in `humidity`:", len(outliers))
```
